```python
spring_df = df[(df['Weather'] == 'spring')]
summer_df = df[df['Weather'] == 'summer']
fall_df = df[df['Weather'] == 'fall']
winter_df = df[df['Weather'] == 'winter']
```

```python
sp = spring_df.groupby(['Company','Capacity']).mean().optimal_rate.unstack()
su = summer_df.groupby(['Company','Capacity']).mean().optimal_rate.unstack()
fa = fall_df.groupby(['Company','Capacity']).mean().optimal_rate.unstack()
wi = winter_df.groupby(['Company','Capacity']).mean().optimal_rate.unstack()
```

```python
sp = spring_df[spring_df['Capacity'] <= 7000].groupby(['Company','Capacity']).mean().optimal_rate.unstack()
su = summer_df[summer_df['Capacity'] <= 7000].groupby(['Company','Capacity']).mean().optimal_rate.unstack()
fa = fall_df[fall_df['Capacity'] <= 7000].groupby(['Company','Capacity']).mean().optimal_rate.unstack()
wi = winter_df[winter_df['Capacity'] <= 7000].groupby(['Company','Capacity']).mean().optimal_rate.unstack()
```

```python
def get_companies(df, n):
    """ 상위, 중간, 하위 n개 회사를 반환 """
    avg_optimal_rate = df.mean(axis=1)  # 각 회사별 평균 optimal_rate를 계산
    top = avg_optimal_rate.nlargest(n)
    bottom = avg_optimal_rate.nsmallest(n)
    middle = avg_optimal_rate.sort_values().iloc[len(df)//2 - n//2 : len(df)//2 + (n+1)//2]
    return top.index.tolist(), middle.index.tolist(), bottom.index.tolist()
```

```python
season_dfs = {
    'spring': sp,
    'summer': su,
    'fall': fa,
    'winter': wi
}

companies_by_season = {
    'top': {},
    'middle': {},
    'bottom': {}
}
```

```python
for season, dd in season_dfs.items():
    top, middle, bottom = get_companies(dd, 5)
    companies_by_season['top'][season] = top
    companies_by_season['middle'][season] = middle
    companies_by_season['bottom'][season] = bottom

company_counts_by_season = {
    'top': {},
    'middle': {},
    'bottom': {}
}

for position in ['top', 'middle', 'bottom']:
    for season in season_dfs.keys():
        company_counts = pd.Series(companies_by_season[position][season]).value_counts().to_dict()
        company_counts_by_season[position][season] = company_counts
```
