```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
matplotlib.rcParams['font.family'] ='Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus'] =False
```

```python
df = pd.read_csv('./paper_table_4.csv')
```

```python
df['benifit'] = df['BeforeOptimalPrice'] - df['BestFitness']
```

```python
df.loc[df['benifit'] <0]
```

```python
df.loc[df['benifit'] < 0].Capacity.value_counts().index
```

```python
data = df.loc[df['benifit'] < 0].Capacity.value_counts()
```

```python
data.values
```

```python
plt.figure(figsize = (10,6))
ax1 = plt.bar(data.index, data.values, width = 700)
plt.bar_label(ax1, fmt = '%d', fontsize = 12)
plt.ylabel('minus count', fontsize = 15)
plt.xlabel('Capacity(Wh)', fontsize = 15)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.show()
```

```python
df.loc[df['benifit'] >0]
```

```python
df['optimal_rate'] = df['benifit'] / df['BeforeOptimalPrice']
```

```python
idx = np.arange(1, 5)
```

```python
plt.figure(figsize=(10, 6))
plt.boxplot([
    df[(df['Capacity'] < 7000) & (df['Weather']=='spring')].optimal_rate*100,
    df[(df['Capacity'] < 7000) & (df['Weather']=='summer')].optimal_rate*100,
    df[(df['Capacity'] < 7000) & (df['Weather']=='fall')].optimal_rate*100,
    df[(df['Capacity'] < 7000) & (df['Weather']=='winter')].optimal_rate*100])
plt.ylim([-5,10])
plt.ylabel('절감율(%)', fontsize = 15)
plt.yticks(fontsize = 12)
plt.xticks(idx, labels=['봄', '여름', '가을', '겨울'], fontsize = 12)
plt.xlabel('계절', fontsize = 15)
plt.show()
```

```python
plot_df = df.loc[df['benifit'] < 0].groupby(['Weather', 'Capacity']).count().optimal_rate.reset_index()
```

```python
plot_df
```

```python
seasons = ['spring', 'summer', 'fall', 'winter']

# 2x2 서브플롯 생성
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

for i, season in enumerate(seasons):
    ax = axs.flatten()[i]
    season_data = plot_df[plot_df['Weather'] == season]
    bars = ax.bar(season_data['Capacity'], season_data['optimal_rate'], width=700)
    ax.set_title(season, fontsize = 15)
    ax.set_xlabel('Capacity', fontsize = 12)
    ax.set_ylabel('Minus Count', fontsize = 12)
    ax.tick_params(axis='both', labelsize=12)

    for bar in bars:
        yval = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2.0, yval, int(yval), ha='center', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
df.loc[df['benifit'] < 0].groupby(['Weather', 'Capacity']).count().optimal_rate.plot(kind='bar', figsize=(12, 6))
plt.xticks(fontsize = 12, rotation = 70)

plt.show()
```
