```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./paper_table_4.csv')
```

```python
df['benifit'] = df['BeforeOptimalPrice'] - df['BestFitness']
```

```python
df.loc[df['benifit'] <0]
```

```python
df.loc[df['benifit'] < 0].Capacity.value_counts().index
```

```python
data = df.loc[df['benifit'] < 0].Capacity.value_counts()
```

```python
data.values
```

```python
plt.figure(figsize = (10,6))
ax1 = plt.bar(data.index, data.values, width = 700)
plt.bar_label(ax1, fmt = '%d', fontsize = 12)
plt.ylabel('minus count', fontsize = 15)
plt.xlabel('Capacity(Wh)', fontsize = 15)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.show()
```
