```python
best_position_list = []
best_fitness_list = []
before_price_list = []
before_optimal_list = []
weather_condition_list = []  # 보유 번호
solar_none_list = eda_df[eda_df['태양광보유']==0].num.unique()


for j in tqdm(range(len(solar_none_list))):
    qq = eda_df[eda_df['num'] == solar_none_list[j]].groupby(['hour']).mean()['전력사용량(kWh)']
    for i in weather_list:
        for k in range(5):
            best_position, best_fitness, before_price, before_optimal = cal_price(ee, qq, i)
            best_position_list.append(best_position)
            best_fitness_list.append(best_fitness)
            before_price_list.append(before_price)
            before_optimal_list.append(before_optimal)
            weather_condition_list.append(i)  

df = pd.DataFrame({
    'weather_condition': weather_condition_list,  
    'best_position': best_position_list,
    'best_fitness': best_fitness_list,
    'before_price': before_price_list,
    'before_optimal': before_optimal_list
})

print(df)
```
