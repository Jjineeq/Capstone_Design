```python
def trade_solar_value(start, end):
        """
        달마다 시작일, 종료일 + 1 일을 int형으로 넣어주시면 됩니다.
        ex) 20230101, 20230132를 넣으면 
        23년 1월 1일 데이터부터 23년 1월 31일 데이터까지가 리턴됩니다.
        하루 데이터는 408건입니다.
        """
        result = pd.DataFrame()
        total_range = pd.date_range(str(start),str(end), freq='d')
        for i in range(len(total_range)):
                url = 'http://apis.data.go.kr/B552115/PvAmountByLocHr/getPvAmountByLocHr'
                params ={'serviceKey' : 'HBsd7IB4qCP0hHVD3lw2kZl+bAP0XU7KmA6WDTB8lh8AAbPtxJmuIgLJne4VTQB6NPgsIf9jRJ6nTjW9ZYykdw==', 
                        'pageNo' : '1', 
                        'numOfRows' : '1000', 
                        'dataType' : 'JSON',
                        'tradeYmd' : str(total_range[i])[0:10].replace('-','')}
                response = requests.get(url, params=params)
                try:
                        data = pd.DataFrame(response.json()['response']['body']['items']['item'])
                        result = pd.concat([result, data])
                except:
                        pass
        df = result
        df.reset_index(inplace=True)
        df = df.iloc[:,1:]
        return df
```
